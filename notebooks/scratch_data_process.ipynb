## Step 1 : Delete non-used cols

```python
## In addition we do the prediction city by city, here we will begin by the "city_id"=1

## duplicate the "raw_data" and do all modif on the duplicated table "data"
data = raw_data.copy()
data.drop(columns=['id', 'category', 'date_last_update', 'source'], inplace=True)
data = data[data['city_id']==1]

## take a look on "data"
data
```

## Step2 : Check the data information and clear duplicated information

```python
## Check all the "type" category
print(np.unique(data.type))
## Check all the "city_id"
print(np.unique(data.city_id))
## Drop all duplicated rows
data = data.drop_duplicates()
## Sort the table by date event
data = data.sort_values(by='date_event')

data
```

## Step3 : Compute Time delta

```python
## Create col "timestamp" by "date_event" for timedelta operation in python
data['timestamp'] = pd.to_datetime(data['date_event'])

## Get "date", "time", "weekday" and "tdelta" information by "timestamp"
data['date'] = data['timestamp'].dt.date
data['time'] = data['timestamp'].dt.time
data['weekday'] = data['timestamp'].dt.strftime('%A')
data['tdelta'] = data['date_event'].diff()
```

```python
np.unique(data.date)
```

```python
## Time interval frequency chart to check abnormal time delta
plt.figure(figsize=(5, 5))
n, bins, patches = plt.hist(data['tdelta'].dropna().dt.total_seconds()/60, bins=20, color='skyblue', edgecolor='black')  # 以min为单位绘制直方图
plt.title('Time interval frequency chart')
plt.xlabel('Timedelta in min')
plt.ylabel('Frequence')
plt.grid(True)

for i in range(len(patches)):
    if int(n[i]) != 0:
        plt.text(patches[i].get_x() + patches[i].get_width() / 2, patches[i].get_height() +30,
                 f'{int(n[i])}', ha='center', va='bottom', c='red')

plt.show()
```

## Step4 : Filter abnormal time delta

```python
data['tdelta_min'] = data['tdelta'].dt.total_seconds() / 60
threshold = 12*60

filtered_data = data[data['tdelta_min'] <= threshold]
filtered_data
```

```python
## Re-check time delta
plt.figure(figsize=(5, 5))
n, bins, patches = plt.hist(filtered_data['tdelta'].dropna().dt.total_seconds()/60, bins=20, color='skyblue', edgecolor='black')  
plt.title('Time interval frequency chart')
plt.xlabel('Timedelta in min')
plt.ylabel('Frequence')
plt.grid(True)

for i in range(len(patches)):
    if int(n[i]) != 0:
        plt.text(patches[i].get_x() + patches[i].get_width() / 2, patches[i].get_height() +30,
                 f'{int(n[i])}', ha='center', va='bottom', c='red')

plt.show()
```
